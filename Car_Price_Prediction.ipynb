```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming 'x' is your DataFrame

# Identify columns with object or string data types
categorical_cols = x.select_dtypes(include=['object', 'string']).columns

# Apply Label Encoding to convert strings to numerical values
label_encoder = LabelEncoder()
for col in categorical_cols:
    x[col] = label_encoder.fit_transform(x[col])

# Now you can proceed with splitting the data and training the model
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)

rf=RandomForestRegressor()
rf.fit(X_train,y_train)

xgb=GradientBoostingRegressor()
xgb.fit(X_train,y_train)

xgbr=XGBRegressor()
xgbr.fit(X_train,y_train)
```

```python
y_pred1=lr.predict(X_test)
y_pred2=rf.predict(X_test)
y_pred3=xgb.predict(X_test)
y_pred4=xgbr.predict(X_test)
```

```python
from sklearn import metrics
score1=metrics.r2_score(y_test,y_pred1)
score2=metrics.r2_score(y_test,y_pred2)
score3=metrics.r2_score(y_test,y_pred3)
score4=metrics.r2_score(y_test,y_pred4)
print(score1,score2,score3,score4)
```

```python
final_data = pd.DataFrame({'Models':['Linear Regression','Random Forest','Gradient Boosting','XGBoost'],'R2 Score':[score1,score2,score3,score4]})
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import joblib

# Load the saved model
model = joblib.load('car_price_prediction')

# Define the Streamlit app
st.title('Car Price Prediction')

# Create user inputs
Present_Price = st.number_input('Present Price of the car (in lakhs)', min_value=0.0, max_value=50.0, value=5.59, step=0.1)
Driven_kms = st.number_input('Kilometers Driven', min_value=0, max_value=500000, value=27000, step=1000)
Fuel_Type = st.selectbox('Fuel Type', options=['Petrol', 'Diesel', 'CNG'])
Selling_type = st.selectbox('Selling Type', options=['Dealer', 'Individual'])
Transmission = st.selectbox('Transmission', options=['Manual', 'Automatic'])
Owner = st.slider('Number of Owners', min_value=0, max_value=3, value=1)
Age = st.slider('Age of the Car (in years)', min_value=0, max_value=30, value=8)

# Map inputs to the correct format for the model
Fuel_Type_map = {'Petrol': 0, 'Diesel': 1, 'CNG': 2}
Selling_type_map = {'Dealer': 0, 'Individual': 1}
Transmission_map = {'Manual': 0, 'Automatic': 1}

# Prepare the input DataFrame for prediction
data_new = pd.DataFrame({
    'Present_Price': [Present_Price],
    'Driven_kms': [Driven_kms],
    'Fuel_Type': [Fuel_Type_map[Fuel_Type]],
    'Selling_type': [Selling_type_map[Selling_type]],
    'Transmission': [Transmission_map[Transmission]],
    'Owner': [Owner],
    'Age': [Age]
})

# Make predictions when the button is clicked
if st.button('Predict Selling Price'):
    prediction = model.predict(data_new)
    st.success(f'Predicted Selling Price:  {prediction[0]:.2f} Lakhs')
```
